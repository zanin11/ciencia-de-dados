# Estimando o modelo de Regress√£o Linear para o consumo

```python
from sklearn.model_selection import train_test_split
```

```python
y = dados.consumo
```

```python
X = dados[['temp_max', 'chuva', 'fds']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 2811)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
modelo.score(X_train, y_train)
```

```python
y_previsto = modelo.predict(X_test)
```

```python
metrics.r2_score(y_test,y_previsto)
```

# Obtendo Previs√µes Pontuais

```python
entrada = X_test[0: 1]
entrada
```

```python
modelo.predict(entrada)
```

```python
temp_max = 20.5
chuva = 20.1
fds = 0
entrada = [[temp_max, chuva, fds]]
modelo.predict(entrada)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
index = ['Intercepto', 'Temperatura Maxima', 'Chuva', 'Fds']
```

```python
pd.DataFrame(data = np.append(modelo.intercept_, modelo.coef_), index = index, columns = ['Parametros'])
```

Intercepto -> Excluindo o efeito das variaveis explicativas X2=X3=X4= 0 o efeito m√©dio no consumo de cerveja seria de 5951.97 L

Temperatura Maxima -> Mantendo-se os valores de Chuva e Fds constantes, o acrescimo de 1¬∫C na Temperatura Maxima gera uma varia√ß√£o m√©dia no consumo de cerveja de 684,74 L

Chuva -> Mantendo-se os valores de Temperatura M√°xima e Fds constantes, o acrescimo de 1 mm de Chuva gera uma varia√ß√£o m√©dia no consumo de cerveja de -60, 78 L

Fds -> Mantendo-se os valores de Temperatura M√°xima e Chuva constantes, o fato do dia ser classificado como final de semana gera uma varia√ß√£o m√©dia no consumo de cerveja de 5401.08 L
